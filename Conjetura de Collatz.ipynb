<h3 align="left"><font color='#DEB887'>ðŸ“Œ Intro:</font></h3>

The Collatz conjecture, also known as the 3n+1 problem or the Syracuse conjecture, is a famous mathematical conjecture formulated as follows:

Given a positive integer n, if it is even, divide it by 2 (n/2), and if it is odd, multiply it by 3 and add 1 (3n+1). The process is repeated with the obtained result, generating a sequence of numbers. The conjecture states that regardless of the initial value of n, this sequence will eventually reach 1.

Mathematically, we can express the process of the Collatz conjecture as follows:

If n is even, n â†’ n/2 If n is odd, n â†’ 3n + 1

This process is iteratively repeated until the value 1 is reached.

The conjecture states that for any positive integer n, regardless of the chosen starting point, the resulting sequence will eventually reach the number 1 and remain in a repetitive cycle of the form (1, 4, 2, 1, 4, 2, ...).

Although the Collatz conjecture has been proven for all integers up to approximately 2^60, its complete proof is still unknown. It remains an open problem in number theory and has intrigued mathematicians for decades.

# <p style="font-family:JetBrains Mono; font-weight:bold; letter-spacing: 2px; color:#DEB887; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #4682B4">First Look</p>

```python
import matplotlib.pyplot as plt

def collatz(n):
    # Create a list to store the iterations
    iterations = [n]
    
    # Perform the iteration until n reaches 1
    while n != 1:
        # If n is even, divide by 2
        if n % 2 == 0:
            n = n // 2
        # If n is odd, multiply by 3 and add 1
        else:
            n = 3 * n + 1
        
        # Add the current number to the list of iterations
        iterations.append(n)
    
    # Return the list of iterations
    return iterations

# Create a list to store the lists of iterations
total_iterations = []

# Iterate over the numbers from 27 to 56 and save the iterations for each number
for i in range(5, 28):
    iterations = collatz(i)
    total_iterations.append(iterations)

# Sort the lists of iterations based on the length of iterations
sorted_iterations = sorted(total_iterations, key=lambda x: len(x), reverse=True)

# Select the top 5 lists with the highest number of iterations
top5_iterations = sorted_iterations[:5]

# Plot each of the lists of iterations in a separate figure
for i, iterations in enumerate(top5_iterations):
    plt.figure(i)
    plt.plot(iterations)
    plt.title(f'Number: {iterations[0]}, Iterations: {len(iterations)}', color='black')

plt.show()
```

```python
data = []

for iteration in top5_iterations:
    row = {
        'Number': iteration[0],
        'Length': len(iteration),
        'Mean': sum(iteration)/len(iteration),
        'Max': max(iteration)
    }
    data.append(row)

df = pd.DataFrame(data)
display(df)
    
```

# <p style="font-family:JetBrains Mono; font-weight:bold; letter-spacing: 2px; color:#DEB887; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #4682B4">Graphs of Collatz Sequences</p>

```python
import networkx as nx

# Create a list to store the lists of iterations
total_iterations = []

# Iterate over the numbers of interest and save the iterations for each number
numbers = [27, 31, 54, 55, 41]
for i in numbers:
    iterations = collatz(i)
    total_iterations.append(iterations)

# Create a directed graph for each list of iterations
for i, iterations in enumerate(total_iterations):
    plt.figure(figsize=(20, 20))
    G = nx.DiGraph()
    for j in range(len(iterations) - 1):
        G.add_edge(iterations[j], iterations[j+1])

    # Define the position of the nodes
    pos = nx.spring_layout(G)

    # Draw the nodes
    nx.draw_networkx_nodes(G, pos=pos, node_size=700, node_color='white', edgecolors='red')

    # Draw the node labels
    labels = {n: n for n in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels, font_size=10, font_family='sans-serif')

    # Draw the edges with white arrows
    edges = G.edges()
    nx.draw_networkx_edges(G, pos=pos, node_size=2000, edgelist=edges, edge_color='blue', arrowsize=10, width=1)

    # Add the title with the number of iterations
    plt.title(f"Number: {numbers[i]}, Iterations: {len(iterations) - 1}", fontsize=12, fontweight='bold', color='black')

    # Show the graph
    plt.axis('off')
    plt.show()
```

# <p style="font-family:JetBrains Mono; font-weight:bold; letter-spacing: 2px; color:#DEB887; font-size:140%; text-align:left;padding: 0px; border-bottom: 3px solid #4682B4">Bentford's Law</p>

```python
def calculate_digit_frequencies(number_list):
    frequencies = {i: 0 for i in range(1, 10)}
    for num in number_list:
        first_digit = int(str(num)[0])
        frequencies[first_digit] += 1
    total_numbers = len(number_list)
    for i in frequencies:
        frequencies[i] /= total_numbers
    return frequencies


initial_number = 10**6
iterations = collatz(initial_number)
digit_frequencies = calculate_digit_frequencies(iterations)

benford_law = {i: np.log10(1 + 1/i) for i in range(1, 10)}

fig, ax = plt.subplots()
ax.bar(benford_law.keys(), benford_law.values(), label='Benford\'s Law')
ax.bar(digit_frequencies.keys(), digit_frequencies.values(), label='Observed Frequency')
ax.set_xticks(list(range(1, 10)))
ax.legend()
plt.title(f"Digit Frequency in the Collatz Conjecture for n={initial_number}")
plt.xlabel('Digits')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate a list of 100 starting numbers
origin_numbers = list(range(1, 101))

# Calculate the second largest number in each Collatz sequence corresponding to the starting numbers
secondary_numbers = []
for n in origin_numbers:
    sequence = list(collatz(n))
    sequence.sort(reverse=True)
    second_largest = None
    for num in sequence:
        if num < n:
            second_largest = num
            break
    secondary_numbers.append(second_largest)

# Plot the results
import matplotlib.pyplot as plt
plt.scatter(origin_numbers, secondary_numbers)
plt.title('Second Largest Number in Collatz Sequence')
plt.xlabel('Starting Number')
plt.ylabel('Second Largest Number')
plt.plot([1, 100], [1, 100], color = "blue")
plt.show()
```
